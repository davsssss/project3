```python
df = pd.read_csv(r'../Resources/Global_Youtube_Statistics-Copy.csv', encoding='ISO-8859-1')
df
```

```python
clean_df = df.dropna()
clean_df
```

```python
# Define a regex pattern to remove special characters
pattern = r'[^a-zA-Z0-9\s]'
draft_df = clean_df.replace(to_replace=pattern, value='', regex=True)
```

```python
file_path = '../Resources/clean_headed_data.csv'
draft_df.to_csv(file_path, index=False)
```
