```python
import pandas as pd
import json
```

```python
df = pd.read_csv('pfc_data.csv')
```

```python
category = df[["Country","Category"]]
category
```

```python
group1 = category.groupby(['Country','Category']).value_counts()
print(group1)
```

```python
result_dict = {}

for index, count in group1.items():
    group, value = index
    if group not in result_dict:
        result_dict[group] = {}
    result_dict[group][value] = count
print(result_dict)
```

```python
json_data = json.dumps(result_dict, indent=4)
js_file_name = "pieChartData.js"
with open(js_file_name, "w") as js_file:
    js_file.write(f"var pieData = {json_data};")
```

```python
channels = df[["Country","Youtuber","Subscribers","Category"]]
channels
```

```python
# Convert each row into a dictionary
row_dicts = channels.to_dict(orient='records')
# Print the list of dictionaries
row_dicts
```

```python
json_data = json.dumps(row_dicts, indent=4)
js_file_name = "barChartData.js"
with open(js_file_name, "w") as js_file:
    js_file.write(f"var barData = {json_data};")
```
